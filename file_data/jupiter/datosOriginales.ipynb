```python
#Cargamos el csv
pichincha = pd.read_csv("/Users/elsa/Desktop/Desafío de tripulaciones/ECUADOR/PICHINCHA/pichincha.csv")
pichincha
```

```python
pichincha.set_index("ESTADO_CONTRIBUYENTE", inplace=True)
```

```python
#Filtro por empresas activas actualmente
pichincha2 = pichincha.loc[["ACTIVO"],:]
pichincha2.reset_index(inplace=True)
```

```python
# Elimino columnas que no nos aportan información
pichincha3 = pichincha2.loc[:,['RAZON_SOCIAL', 'NOMBRE_COMERCIAL', 'CALLE', 'NUMERO', 'INTERSECCION', 'DESCRIPCION_PROVINCIA',"CODIGO_CIIU", 'DESCRIPCION_CANTON','ACTIVIDAD_ECONOMICA']]
```

```python
pichincha3.isnull().sum()
```

```python
# Convertimos los NaN en Null porque lenguaje sql que es donde se van a exportar estos datos NaN = Null
pichincha3['NOMBRE_COMERCIAL'].fillna("Null", inplace=True)
pichincha3
```

```python
# Obtención de todas las actividades económicas que son Length: 1476
for elem in pichincha3.ACTIVIDAD_ECONOMICA:
    lambda elem: elem.startswith('CULTIVO')
    print(elem)
```

```python
# Obtención de una lista donde estén todas las actividades económicas para poder filtrar las actividades en nuestros datos.
lista = []
for elem in pichincha3.ACTIVIDAD_ECONOMICA:
    if isinstance(elem, float) == False:
        if elem.startswith('CULTIVO') == True:
            lista.append(elem)
print("Número de actividades: ",len(set(lista)))
print("---------------------------")
set(lista)
```

```python
# Filtrar todo lo que no esté en la lista
pichincha4 = pichincha3[pichincha3['ACTIVIDAD_ECONOMICA'].isin(lista)]
pichincha4
```
